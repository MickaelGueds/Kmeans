**Imports**

```python
from sklearn.cluster import KMeans
import numpy as np
```

O que vamos começar fazendo, vamos treinar com dados aleatorios,o uso do fit_preditc serve pra treinar o modelo ajustando os centroides para os dados em especificos,depois dissoeles recebem um **rotulo** que indica a qual rotulo ele pertence, de 0 ao 4 -> (5 clusters)

```python
k = 5
kmeans = KMeans(n_clusters = k)
X = np.random.rand(100, 2)
y_pred = kmeans.fit_predict(X)
y_pred
```

```python
y_pred is kmeans.labels_
```

Esse metodo aqui é para mostrar, quais são os centroides de cada kluster, como são dados com duas coluans ele mostra em versão de centroides

```python
kmeans.cluster_centers_
```

Ao usar o kmeans.predict lembre-se que quando realizamos o treino la do inicio, o metodo fit_predit armazena o treinamento no objeto kmeans que tem 5 clusters, logo ao usar esse objeto e outro metodo e atribui-lo em uma variavel isso acontece, o foco é o objeto kmeans

```python
X_new = np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
kmeans.predict(X_new)
```

Agora vamos usar o metodo transform(), que para cada instancia cada ponto que vai ser representado por linha e cada kluster por coluna, vai ser mostrado a distancia de cada instancia para cada centroide, veja, e compare com o preditc de X_new, veja que a coluna com menor valor fica resultante com o valor estabelecido de cada predição, são complementos para uma melhor vizualização

```python
print(kmeans.transform(X_new))
print(kmeans.predict(X_new))
```

E em caso da randomização dos kluster não ser efetiva?
Veja o exemplo logo abaixo de clusterização que deu *Errado* ,bom uma das soluções que o livro aponta seria a utilização de um cluster anterior, caso voce tenha usado outro algoritimo de cluster e voce **Sabe onde aproximadamentes os clusters estão**




```python
good_init = np.array([[-3,3],[-3,2],[-3,1],[-1,2],[0,2]]) #metodo para a criação do array(tipo de objeto que permite uma manipulação melhor dos dado quando se trata de menor consumo de memoria e vetorização matematica para o array que tiver um unico tipo)
kmeans = KMeans(n_clusters=5,init=good_init,n_init=1)
```
